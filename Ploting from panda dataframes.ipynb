### things to do in price column
1) remove the dollar sign
2) the dot 
3) the zeros after dots
4) convert the string int for plot

```python
type(car_sales["Price"])
```

```python
# where regrex is regular explression for matching string
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Price"]= car_sales["Price"].astype(int)
car_sales["Price"].dtypes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
heart_record = pd.read_csv("Data/heart-disease.csv")
len(heart_record)
```

```python
heart_record.head()
```

```python
heart_record.plot(figsize=(10,5) );
```

```python
heart_record.plot.hist()
```

the above are the meaningless ploting

two kind of plotting => pyplot and matplotlib OO method (object oriented method)
1) pyplot for smthing simple
2) matplotlib OO for smthing complex

```python
heart_record.head()
```

```python
# ploting data over 50 year age and male people
over50_maledata = heart_record[(heart_record["age"] > 50) & (heart_record["sex"] == 1)]
over50_maledata
```

```python
#pyplot method 
over50_maledata.plot(kind="scatter",
                    x= "chol", y= "age",c="target",ylabel ="age of male");
```

```python
# 00 method, how to do it in pandas dataframe , is one way of oo method
fig , ax = plt.subplots()
over50_maledata.plot(kind="scatter" , x="age",y="chol",c="target",ax =ax);
```

```python
over50_maledata[:50]
```

```python
#other way oo method
# let's see from scr`atch

fig, ax = plt.subplots(figsize=(10,10))

# plot the data
scatter = ax.scatter(x=over50_maledata["age"],
                          y = over50_maledata["chol"],
                        c= over50_maledata["target"] )

# customize the data
ax.set(title="heart disease and cholestrol level with age",ylabel="cholestrol--", xlabel="age--")
ax.plot()

# add the legend ( to know what is happening, which color is repesenting what )
ax.legend(*scatter.legend_elements(),title="yellow represent the heart disease",loc="upper right");

# # optional , add meanline
# ax.axhline(y=over50_maledata["chol"].mean(),linestyle ="--");
```
