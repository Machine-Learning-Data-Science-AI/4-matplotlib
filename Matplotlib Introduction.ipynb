so, in matplotlib be have two types of ploting:
1) implicit Pyplot API  = The implicit Pyplot API is a high-level interface for creating Matplotlib plots quickly and with less code. it's disadvantage is , it is less flexible while plot creation and modification.
2) explicit Pyplot API / object oriented = The explicit API is a lower-level interface that gives you more control and flexibility over the details of your Matplotlib plots.

## Implicit pyplot api

```python
x =[23, 45,67 ,100]
y=[1,2,3,90]
plt.plot(x,y)
plt.title("this is implicit pyplot api")
plt.xlabel("x-axis");
plt.ylabel("y-axis");
```

#### workflow of explicit pyplot api

```python
# 0. import the matplotlib 
%matplotlib inline
import matplotlib.pyplot as plt 

# 1. prepare the data
x=[2,3,4]
y=[23,45,0]

# 2. setup the plot
fig , ax = plt.subplots(figsize = (10 ,10))

# 3. plot the data
ax.plot(x,y)

# 4. customize plort
ax.set(title="explicit pyplot api",
       xlabel="x-axis",
      ylabel="y-axis");

# 5. save and show yr work
fig.savefig("sample-plot3.png")
```
