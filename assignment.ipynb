```python
heart_disease =  heart_disease[heart_disease["age"]>50] 
len(heart_disease)
```

```python
heart_disease.head()
```

```python
fig , axis = plt.subplots()

scatter = axis.scatter(x=heart_disease["age"] , y=heart_disease["chol"],c=heart_disease["target"])

axis.set(title="Heart Disease and Cholesterol Levels",xlabel="age",ylabel="chol")

axis.legend(*scatter.legend_elements() ,title="yellow represent the heart disease", loc="upper right");


axis.axhline(y=heart_disease["chol"].mean(),linestyle ="--");
```
