```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Scatter graph


```python
fig , ax = plt.subplots()
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
ax.scatter(x,y);
# or
# ax.scatter([5,7,8,7,2,17,2,9,4,11,12,9,6],[99,86,87,88,111,86,103,87,94,78,77,85,86]);
```

## Bar graph

```python
fig , ax = plt.subplots()
x = np.array(["india","canada","us"])
y =np.array([60,20,80])
ax.bar(x,y)
ax.set(label ="country progress per year report" , xlabel="countries name----",ylabel="progress in percentage-----");
```

```python
fig , ax = plt.subplots()
x = np.array(["india","canada","us"])
y =np.array([60,20,80])
ax.barh(x,y)
ax.set(label ="country progress per year report" , ylabel="countries name----",xlabel="progress in percentage-----");
```

## multiploting 

#### option 1

```python
# 2 rows and colms making the matrix , two rows and column
fig , ax = plt.subplots(nrows=2,
                        ncols=2
                       )
ax[0,0].plot([1,2,3],[23,45,1]);
ax[0,1].scatter([1,2,3],[23,45,1]);
ax[1,0].bar(["a","b","c"],[23,43,56]);
ax[1,1].hist([1,1,1,1,2,2,2,45,45,45,45]);
```

#### option2

```python
# 3 rows and 2 colms making the matrix , three rows and two column
fig , ((ax0 , ax1),(ax2,ax3),(ax4,ax5)) = plt.subplots(nrows=3,
                        ncols=2
                       )
ax0.plot([1,2,3],[23,45,1]);
ax2.scatter([1,2,3],[23,45,1]);
ax3.bar(["a","b","c"],[23,43,56]);
ax5.hist([1,1,1,1,2,2,2,45,45,45,45]);
```
